```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("../data/raw/DataCoSupplyChainDataset.csv", encoding="latin1")

print("Dataset Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
df.describe()
```

```python
columns_to_drop = [
    "Product Description",
    "Order Zipcode",
    "Customer Email",
    "Customer Password",
    "Customer Fname",
    "Customer Lname",
    "Customer Street",
    "Product Image"
]

df.drop(columns=columns_to_drop, inplace=True)
print("New Shape:", df.shape)
```

```python
df["order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"])
df["shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"])
```

```python
df.dtypes
```

```python
df.to_csv("../data/processed/cleaned_supply_chain.csv", index=False)
```

```python
# On-time delivery rate
on_time_rate = (df["Late_delivery_risk"] == 0).mean() * 100
print("On-Time Delivery Rate:", round(on_time_rate, 2), "%")

# Late delivery rate
late_rate = (df["Late_delivery_risk"] == 1).mean() * 100
print("Late Delivery Rate:", round(late_rate, 2), "%")
```

```python
# Actual vs Scheduled difference
df["Shipping_Delay"] = df["Days for shipping (real)"] - df["Days for shipment (scheduled)"]

print("Average Shipping Delay:", round(df["Shipping_Delay"].mean(), 2), "days")
```

```python
df["Shipping_Delay"].describe()
```

```python
# Late rate by Shipping Mode
shipping_mode_analysis = df.groupby("Shipping Mode")["Late_delivery_risk"].mean() * 100
print(shipping_mode_analysis.sort_values(ascending=False))
```

```python
region_analysis = df.groupby("Order Region")["Late_delivery_risk"].mean() * 100
print(region_analysis.sort_values(ascending=False))
```

```python
# Profit by Shipping Mode
profit_mode = df.groupby("Shipping Mode")["Order Profit Per Order"].mean()
print(profit_mode.sort_values(ascending=False))
```

```python
# Profit by Region
profit_region = df.groupby("Order Region")["Order Profit Per Order"].mean()
print(profit_region.sort_values(ascending=False).head(10))
```

```python
total_sales = df["Sales"].sum()
total_profit = df["Order Profit Per Order"].sum()
avg_profit = df["Order Profit Per Order"].mean()
profit_margin = (total_profit / total_sales) * 100
```
